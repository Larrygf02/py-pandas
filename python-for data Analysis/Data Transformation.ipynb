## Data using a function or mapping

```python
data = DataFrame({'food': ['bacon', 'pulled pork', 'bacon', 'Pastrami',
        'corned beef', 'Bacon', 'pastrami', 'honey ham',
        'nova lox'],
        'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
meat_to_animal = {
 'bacon': 'pig',
 'pulled pork': 'pig',
 'pastrami': 'cow',
 'corned beef': 'cow',
 'honey ham': 'pig',
 'nova lox': 'salmon'
}
data['animal'] = data['food'].map(str.lower).map(meat_to_animal)
```

```python
data
```

```python
## Same example with person and genrs
df1 = DataFrame({'name': ['raul', 'grace', 'ernesto', 
                          'yeli', 'helen', 'josue'],
                'edad': [26,26,32,32, 60, 20]})
df1
```

```python
meat_to_genre = {
    'raul': 'male',
    'grace': 'female',
    'ernesto': 'male',
    'yeli': 'female',
    'helen': 'female',
    'josue': 'male'
}
df1['genre'] = df1['name'].map(str.lower).map(meat_to_genre)
df1
```

```python
group_ages = [18, 25, 40, 60]
person_by_group = pd.cut(df1['edad'], group_ages)
pd.value_counts(person_by_group)
```
